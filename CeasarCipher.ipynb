**Ceasar Cipher Encryption Part**

```python
def encrypt_caesar(plain_text, shift):
    encrypted_text = ""
    for char in plain_text:
        if char.isalpha():
            # Determine whether the character is uppercase or lowercase
            is_upper = char.isupper()

            # Apply the shift to the character
            char_code = ord(char) + shift
            # Wrap around the alphabet
            if is_upper:
                char_code = (char_code - ord('A')) % 26 + ord('A')
            else:
                char_code = (char_code - ord('a')) % 26 + ord('a')

            encrypted_text += chr(char_code)
        else:
            encrypted_text += char
    return encrypted_text

# Example Usage:
plaintext = "Hello, Hilal!"
shift_amount = 6
cipher_text = encrypt_caesar(plaintext, shift_amount)
print(f"Encrypted Text: {cipher_text}")
```

**Ceasar Cipher Decryption Part**

```python
def decrypt_caesar(cipher_text, shift):
    decrypted_text = ""
    for char in cipher_text:
        if char.isalpha():
            # Determine whether the character is uppercase or lowercase
            is_upper = char.isupper()

            # Apply the reverse shift to the character
            char_code = ord(char) - shift
            # Wrap around the alphabet
            if is_upper:
                char_code = (char_code - ord('A')) % 26 + ord('A')
            else:
                char_code = (char_code - ord('a')) % 26 + ord('a')

            decrypted_text += chr(char_code)
        else:
            decrypted_text += char
    return decrypted_text

# Example Usage:
decrypted_text = decrypt_caesar(cipher_text, shift_amount)
print(f"Decrypted Text: {decrypted_text}")
```
